```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("ds_salaries.csv")
```

```python
df.head(2)
```

```python
# checking for null values
df.isnull().any()
```

From above results we can tell that the data does not contain any null values

Their is one column which does not have the header and it is not useful for the analysis

```python
# droping the Unnamed column
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.shape
```

```python
# checking the count of unique values in each column
df.nunique()
```

```python
# ploting a graph that shows the count of the work years
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='work_year',width=0.6)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above result we can see that most of the job are in 2022

```python
# ploting a graph that shows the count of the Experience level
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='experience_level',width=0.6)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# ploting a graph that shows the count of the Job Titles
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='job_title')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 90)
plt.show()
```

```python
# ploting a graph that shows the count of the salary currency
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='salary_currency')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# ploting a graph that shows the count of the Employment Type
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='employment_type',width=0.6)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above result we can see that most of the Jobs are prefered Full Time Employees

```python
# ploting a graph that shows the count of the Company Location
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='company_location')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above result we can observe that the Jobs are created in US 

```python
# ploting a graph that shows the count of the Company Size
plt.figure(figsize = (20,10))
ax = sns.countplot(data = df,x='company_size',width=0.6)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Most Jobs are provided by the Medium size company

### Analysis based on Average Salary

```python
# calculating average salary based on employment type
avg_emp_type = df.groupby('employment_type')['salary'].mean().reset_index()
```

```python
print(avg_emp_type)
```

```python
plt.figure(figsize = (20,10))
ax=sns.barplot(data = avg_emp_type,x='employment_type',y='salary')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above visual we can observe that fulltime employees have the highest Average salary

```python
# calculating average salary based on the Job Title
avg_job_title = df.groupby('job_title')['salary'].mean().reset_index()
```

```python
print(avg_job_title)
```

```python
plt.figure(figsize = (20,10))
ax=sns.barplot(data = avg_job_title,x='job_title',y='salary')
for bars in ax.containers:
    ax.bar_label(bars,rotation = 90,fmt='%2f',padding=1)
plt.xticks(rotation = 90)
plt.show()
```

```python
# calculating average salary based on Experience Level
avg_exp_level = df.groupby('experience_level')['salary'].mean().reset_index()
```

```python
print(avg_exp_level)
```

```python
plt.figure(figsize = (20,10))
ax=sns.barplot(data = avg_exp_level,x='experience_level',y='salary')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# overall Average salary in each year
avg_work_year = df.groupby('work_year')['salary'].mean().reset_index()
```

```python
print(avg_work_year)
```

```python
plt.figure(figsize = (20,10))
ax=sns.barplot(data = avg_work_year,x='work_year',y='salary')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above visual we can understand that in 2021 the overall average salary is highest

```python
# calculating Average salary based on company size
avg_comp = df.groupby('company_size')['salary'].mean().reset_index()
```

```python
print(avg_comp)
```

```python
plt.figure(figsize = (20,10))
ax=sns.barplot(data = avg_comp,x='company_size',y='salary')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above visual we can easily understand that highest average salary is in large companies
